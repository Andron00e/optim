```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the norms
def p1_norm(x, y):
    return np.abs(x) + np.abs(y)

def p2_norm(x, y):
    return np.sqrt(x**2 + y**2)

def pinf_norm(x, y):
    return np.maximum(np.abs(x), np.abs(y))

# Generate grid
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z1 = p1_norm(X, Y)
Z2 = p2_norm(X, Y)
Z3 = pinf_norm(X, Y)

# Plotting
fig = plt.figure(figsize=(12, 4))

# Plot for p=1 norm
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z1, alpha=0.8, cmap='coolwarm',  edgecolor='none')
ax1.set_title('$p=1$ Norm Cone')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('$t$')
ax1.xaxis.pane.fill = False
ax1.yaxis.pane.fill = False
ax1.zaxis.pane.fill = False
ax1.xaxis.pane.set_edgecolor('w')
ax1.yaxis.pane.set_edgecolor('w')
ax1.zaxis.pane.set_edgecolor('w')
ax1.xaxis._axinfo["grid"]['linestyle'] = ":"
ax1.yaxis._axinfo["grid"]['linestyle'] = ":"
ax1.zaxis._axinfo["grid"]['linestyle'] = ":"

# Plot for p=2 norm
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X, Y, Z2, alpha=0.8, cmap='coolwarm',  edgecolor='none')
ax2.set_title('$p=2$ Norm Cone')
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_zlabel('$t$')
ax2.xaxis.pane.fill = False
ax2.yaxis.pane.fill = False
ax2.zaxis.pane.fill = False
ax2.xaxis.pane.set_edgecolor('w')
ax2.yaxis.pane.set_edgecolor('w')
ax2.zaxis.pane.set_edgecolor('w')
ax2.xaxis._axinfo["grid"]['linestyle'] = ":"
ax2.yaxis._axinfo["grid"]['linestyle'] = ":"
ax2.zaxis._axinfo["grid"]['linestyle'] = ":"

# Plot for p=inf norm
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(X, Y, Z3, alpha=0.8, cmap='coolwarm',  edgecolor='none')
ax3.set_title('$p=\infty$ Norm Cone')
ax3.set_xlabel('$x$')
ax3.set_ylabel('$y$')
ax3.set_zlabel('$t$')
ax3.xaxis.pane.fill = False
ax3.yaxis.pane.fill = False
ax3.zaxis.pane.fill = False
ax3.xaxis.pane.set_edgecolor('w')
ax3.yaxis.pane.set_edgecolor('w')
ax3.zaxis.pane.set_edgecolor('w')
ax3.xaxis._axinfo["grid"]['linestyle'] = ":"
ax3.yaxis._axinfo["grid"]['linestyle'] = ":"
ax3.zaxis._axinfo["grid"]['linestyle'] = ":"

plt.tight_layout()
plt.savefig("norm_cones.pdf")
plt.show()
```
