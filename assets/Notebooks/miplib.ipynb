```python
import pandas as pd
import matplotlib.pyplot as plt

# URL to the CSV file
url = "http://fmin.xyz/assets/files/miplib_collection.csv"

# Download the CSV file and load it into a DataFrame
df = pd.read_csv(url)

# Define colors and markers for different statuses
color_map = {
    'easy': ('green', 'o'),  # green circle
    'hard': ('yellow', 'D'),  # yellow diamond
    'open': ('red', 's')   # red square
}

true_label = {
    'easy': "Easy (< 1 hour)", 
    'hard': "Hard (> 1 hour)",
    'open': "Unsolved"
}

# Create the scatter plot
plt.figure(figsize=(7, 5))
for status, group in df.groupby('StatusStat.'):
    plt.scatter(group['VariablesVari.'], group['ConstraintsCons.'], 
                color=color_map[status][0], marker=color_map[status][1], label=true_label[status], edgecolors='k', zorder=3)

plt.xlabel('Number of Variables')
plt.ylabel('Number of Constraints')
plt.xscale("log")
plt.yscale("log")
plt.title('Running time to optimality for different MIPs.\nMIPLIB 2017 Collection Set')
plt.legend()
plt.grid(linestyle=":", zorder=0)
plt.savefig("miplib.pdf")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URL to the CSV file
url = "http://fmin.xyz/assets/files/miplib_collection.csv"

# Download the CSV file and load it into a DataFrame
df = pd.read_csv(url)

# Define colors and markers for different statuses
color_map = {
    'easy': ('green', 'o'),  # зеленый круг
    'hard': ('yellow', 'D'),  # желтый ромб
    'open': ('red', 's')   # красный квадрат
}

true_label = {
    'easy': "Легкая (< 1 часа)", 
    'hard': "Тяжелая (> 1 часа)",
    'open': "Не решена"
}

# Create the scatter plot
plt.figure(figsize=(7, 5))
for status, group in df.groupby('StatusStat.'):
    plt.scatter(group['VariablesVari.'], group['ConstraintsCons.'], 
                color=color_map[status][0], marker=color_map[status][1], label=true_label[status], edgecolors='k', zorder=3)

plt.xlabel('Количество переменных')
plt.ylabel('Количество ограничений')
plt.xscale("log")
plt.yscale("log")
plt.title('Время решения для различных задач MIP.\nДатасет MIPLIB 2017 Collection Set')
plt.legend()
plt.grid(linestyle=":", zorder=0)

# Add watermark at the bottom right corner
plt.text(1, -0.13, '@fminxyz', fontsize=10, color='grey', ha='right', va='bottom', alpha=0.5, transform=plt.gca().transAxes)

plt.savefig("miplib_ru.pdf")
plt.show()
```
