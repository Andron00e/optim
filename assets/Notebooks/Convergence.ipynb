```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

C = 10
alpha = -7
q = 0.5
quad_q = 0.9
supl_q = 0.9
num_iter = 50
sublinear = np.array([C * k**alpha for k in range(1, num_iter + 1)])
linear = np.array([C * q**k for k in range(1, num_iter + 1)])
superlinear = np.array([C * supl_q**(k**2) for k in range(1, num_iter + 1)])
quadratic = np.array([C * quad_q**(2**k) for k in range(1, num_iter + 1)])
plt.figure(figsize=(5,3))
plt.semilogy(np.arange(1, num_iter+1), sublinear, 
             label=f"Sublinear, α = {alpha}")
plt.semilogy(np.arange(1, num_iter+1), superlinear, 
             label=f"Superlinear, q = {supl_q}")
plt.semilogy(np.arange(1, num_iter+1), linear, 
             label=f"Linear, q = {q}")
plt.semilogy(np.arange(1, num_iter+1), quadratic, 
             label=f"Quadratic, q = {quad_q}")
plt.xlabel("Number of iteration, $k$")
plt.ylabel("Error rate upper bound")
plt.ylim(1e-12, 2e1)
plt.legend(loc="best")
plt.grid(linestyle=":")
plt.tight_layout()
plt.savefig('convergence.svg')
```
